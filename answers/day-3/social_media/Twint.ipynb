```python
import random
from time import sleep


keywords = ['covid', 'pandemic', 'depression']

for _keywords in keywords:
    counter = 0
    while counter < 3:
        c = twint.Config()
        c.Search = _keywords
        c.Store_object = True
        c.Lang="en"
        c.Since="2021-01-01"
        c.Until="2021-12-31"
        c.Geo="1.290270,103.851959,30km"
        c.Limit = 3000
        c.Store_csv = True
        c.Output = f"twint-{_keywords}.csv"
        result = twint.run.Search(c)

        if result is None:
            counter +=1
            sleep(random.randint(5, 10))
        else:    
            break
```
